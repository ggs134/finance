```python
import pandas as pd
```

```python
import pymongo
```

```python
cli = pymongo.MongoClient('192.168.1.19',27017)
db = cli.ticker
```

```python
data = db.btce.find()
```

```python
data[0]
```

```python
time=[]
korbit = []
btce = []

for i in data:
        time.append(i['time'])
        btce.append(i['BTCe']['btc_usd']['avg'])
        korbit.append(i['Korbit']['last'])
```

```python
len(time)
```

```python
len(korbit)
```

```python
len(btce)
```

```python
frame = pd.DataFrame(korbit, columns=['korbit'], index=time)
```

```python
frame
```

```python
frame['btce']=btce
```

```python
frame
```

```python
exrate = 1182
```

```python
f = lambda x : x*exrate
```

```python
trans_btce = frame['btce'].apply(f)
```

```python
frame['t_btce']=trans_btce
```

```python
frame['korbit']=frame['korbit'].apply(lambda x : float(x))
```

```python
frame['diff']=frame['korbit']-frame['t_btce']
```

```python
frame['spread']=(frame['diff'].apply(lambda x : x*100)/frame['korbit'])
```

```python
frame2 = frame[frame['spread']>5]
```
